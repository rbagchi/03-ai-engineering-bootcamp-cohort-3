### Import Dependencies

```python
from langsmith import Client
```

#### Attach Feedback to a specific trace in LangSmith

019c199e-45d3-7102-98d0-fce813e36ae7

```python
client = Client()
```

```python
RUN_ID = "019c199e-45d3-7102-98d0-fce813e36ae7"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by Aurimas",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by Aurimas",
    score=0,
    feedback_source_type="api"
)
```
